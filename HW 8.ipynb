### Chapter 7, Problem 4

Probability of star-star lensing in the Galactic disk

Distance to galactic bulge $d = 8\ kpc$, disk height $h_{disk} = 1\ kpc$, typical star mass $M = 0.5M_\odot$, mass interior to solar radius $M_{d<d_\odot} = 5\times10^{10} M_\odot$

The probability of a gravitational lensing event by a star halfway between the Earth and the bulge is given roughly by

$$ P \sim N \sigma l $$

The number density of stars $N$ can be calculated using the given masses and dimensions of the galaxy:

$$ N = \frac{M_{d<d_\odot}}{M_\odot \pi r^2 h_{disk}} = \frac{5\times10^{10}}{0.5 \pi (8\ kpc)^2 \cdot 1\ kpc} $$
 
$$ N = 4.97\times10^8\ kpc^{-3} $$

The effective cross section of a lensing event is the area of the Einstein ring formed from the lens. The radius of the Einstein ring is calculated as (using the approximation $\tan \theta \approx \theta$)

$$ \tan \theta_E \approx \theta_E = \frac{R_E}{D_{ol}} $$

$$ \sigma = \pi R_E^2 = \pi \left( D_{ol} \theta_E \right)^2 $$

Substituting the formula for the Einstein angle,

$$ \theta_E = \left( \frac{4GM}{c^2} \frac{D_{ls}}{D_{ol} D_{os}} \right)^{1/2} $$

$$ \sigma = \pi D_{ol}^2 \left( \frac{4GM}{c^2} \frac{D_{ls}}{D_{ol} D_{os}} \right) = \frac{4GM}{c^2} \frac{D_{ls} D_{ol}}{D_{os}} $$

$$ \sigma = 6.035\times10^{-16}\ kpc^{-2} $$

Taking $l$ as the bulge distance of 8 kpc, the probability of a lensing event from a star halfway to the Galactic center is

$$ P \sim N \sigma l = 4.97\times10^8\ kpc^{-3} \cdot 6.035\times10^{-16}\ kpc^{-2} \cdot 8\ kpc $$

$$ P \sim 2.401\times10^{-6} $$

### Chapter 8, Problem 3

Sunayev-Zeldovich Effect, 0.001 photons scattered through cluster, $M_{cluster} = 2\times10^{14}M_\odot$, 15% of mass is ionized hydrogen

a. The Compton scattering coefficient $\gamma$ is calculated as

$$ \gamma = \frac{\sigma_T}{m_e c^2} \int n_e(r) k_e T_e(r) dx $$

For a cluster with constant density and temperature, $n_e$ and $T_e$ are constant and the expression for $\gamma$ becomes

$$\gamma = \frac{\sigma_T}{m_e c^2} n_e k_e T_e x = \frac{k_e T_e}{m_e c^2}\ n_e \sigma_T x$$

The quantity $n_e \sigma_T x$ is the probability of an electron upscatter due to the Sunayev-Zeldovich effect, which is given as 0.001. This can be reversed to give the diameter of the cluster $d$.

The number density $n_e$ is the same as the ionized hydrogen density $n_H$, and is caculated by dividing the total mass of ionized hydrogen by the mass of hydrogen times the cluster's volume.  Since only 15% of the cluster's mass is ionized, the total hydrogen mass is 15% of the total cluster mass.

$$ n_e = n_H = \frac{0.15M_{cluster}}{m_H \frac{4}{3} \pi \left( \frac{d}{2} \right)^3 } $$

The cross-section $\sigma$ is the Thomson cross section for electron scattering:

$$ \sigma = \frac{8 \pi}{3} \left( \frac{e^2}{m_e c^2} \right)^2 $$

Combining these into the probability expression, with the distance $x$ equaling the cluster diameter $d$,

$$ P = 0.001 = \frac{0.15M_{cluster}}{m_H \frac{4}{3} \pi \left( \frac{d}{2} \right)^3 } \cdot \frac{8 \pi}{3} \left( \frac{e^2}{m_e c^2} \right)^2 \cdot d $$

$$ 0.001 = 2.4 \frac{M_{cluster}}{m_H} \left( \frac{e^2}{m_e c^2} \right)^2 \frac{1}{d^2} $$

$$ d = \sqrt{2400 \frac{M_{cluster}}{m_H}} \frac{e^2}{m_e c^2} $$

$$ d = 6.685\times10^{24}\ cm = 2.17\ Mpc $$
