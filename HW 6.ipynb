### Problem 5
Strong shocks $(P_1 >> P_0)$, $\gamma = \frac{5}{3}$, $E_{k,0} = \frac{v_0^2}{2}$

a. Thermal energy per unit mass $\epsilon_1$ (shocked gas)

The thermal energy $\epsilon_1$ can be expressed as

$$\epsilon_1 = \frac{1}{\gamma - 1} \frac{P_1}{\rho_1}$$

This makes the ratio of energies

$$\frac{\epsilon_1}{E_{k,0}} = \frac{2}{\gamma - 1}\frac{P_1}{\rho_1 v_0^2} = \frac{2}{\gamma \left(\gamma - 1\right)} \left(\frac{\gamma P_1}{\rho_1}\right) \frac{1}{v_0^2}$$

This expression can be simplified using the equation for sound speed $c_s = \sqrt{\frac{\gamma P}{\rho}}$ and the mass continuity relation $\rho_0 v_0 = \rho_1 v_1$ to get

$$\frac{\epsilon_1}{E_{k,0}} = \frac{2}{\gamma \left(\gamma - 1\right)} \left(\frac{\rho_0}{\rho_1}\right)^2 \frac{c_{s,1}^2}{v_1^2}$$

This can be further simplified using the Mach number $M_1^2 = \left( \frac{v_1}{c_{s,1}} \right)^2 = \frac{\gamma + 1}{2 \gamma} \frac{P_1}{P_0}$, as well as the density relation $\frac{\rho_1}{\rho_0} = \frac{\gamma + 1}{\gamma - 1}$

$$\frac{\epsilon_1}{E_{k,0}} = \frac{2}{\gamma \left(\gamma - 1\right)} \left(\frac{\gamma - 1}{\gamma + 1}\right)^2 \frac{2 \gamma}{\gamma + 1} \frac{P_0}{P_1} = \frac{4 (\gamma - 1)}{(\gamma + 1)^3} $$

Using the value of $\gamma = \frac{5}{3}$,

$$\frac{\epsilon_1}{E_{k,0}} = \frac{4(2/3)}{(8/3)^3} = \frac{9}{64}$$
***



b. Kinetic energy per mass after the shock, in reference of the shock

The kinetic energy of the shock is

$$E_{k,1} = \frac{v_1^2}{2}$$

so the ratio of the two energies is

$$\frac{E_{k,1}}{E_{k,0}} = \frac{v_1^2}{v_0^2} = \left( \frac{v_1}{v_0} \right)^2$$

This ratio of velocities can be related to the ratio of densities (and the adiabatic constant $\gamma$) by

$$\frac{\rho_1}{\rho_0} = \frac{v_0}{v_1} = \frac{\gamma + 1}{\gamma - 1}$$

$$\frac{E_{k,1}}{E_{k,0}} = \left( \frac{v_1}{v_0} \right)^2 = \left( \frac{\gamma - 1}{\gamma + 1} \right)^2$$

Using the value of $\gamma = \frac{5}{3}$,

$$\frac{E_{k,1}}{E_{k,0}} = \left( \frac{2/3}{8/3} \right)^2$$

$$\frac{E_{k,1}}{E_{k,0}} = \frac{1}{16}$$
***



c. Kinetic energy of shocked gas, in reference of the observer

In the observer's reference frame, the post-shock gas moves at a velocity $v = v_0 - v_1$. Thus, the kinetic energy is

$$E_{k,1} = \frac{v^2}{2} = \frac{\left(v_0 - v_1\right)^2}{2}$$

This makes the ratio of energies

$$\frac{E_{k,1}}{E_{k,0}} = \frac{\left(v_0 - v_1 \right)^2}{v_0^2} = \left( \frac{v_0 - v_1}{v_0} \right)^2 = \left( 1 - \frac{v_1}{v_0} \right)^2$$

From part b, $\frac{v_1}{v_0} = \frac{\gamma - 1}{\gamma + 1}$, so using $\gamma = \frac{5}{3}$,

$$\frac{E_{k,1}}{E_{k,0}} = \left( 1 - \frac{\gamma - 1}{\gamma + 1} \right)^2 = \left( 1 - \frac{2/3}{8/3} \right)^2 = \left( \frac{3}{4} \right)^2$$

$$\frac{E_{k,1}}{E_{k,0}} = \frac{9}{16}$$

### Problem 6

SNR1006, forward shock with velocity $v_0 = 6000\ km\ s^{-1}$, shock rim thickness $d = 0.1\ pc$.

Electrons are hypothesized to lose energy to synchrotron radiation at the shock front, producing synchrotron emission at the rim with $h\nu_{sync} = 4\ keV$.

a.  If the electrons are moving with the post-shock material, they have a velocity of $v_e = \frac{3}{4}v_0 = 4500\ km\ s^{-1}$.  At this speed, they cross the 0.1pc shock rim in a time $t_{cross} \sim \frac{d}{v_e}$.

Equating this time to the cooling time $t_{c} = \frac{m_e^3 c^5}{e^4} \frac{1}{\gamma B^2}$,

$$ \frac{d}{v_e} = \frac{m_e^3 c^5}{e^4} \frac{1}{\gamma B^2} $$

This equation can be solved for $\gamma B^2$ to get

$$ \gamma B^2 = \frac{m_e^3 c^5 v_e}{e^4 d} = \frac{\left(9.1\times10^{-28}\right)^3 \left(3\times10^{10}\right)^5 \left(4500\times10^5\right)}{\left(4.8\times10^{-10}\right)^4 \left(3.1\times10^{17}\right)}$$

$$ \gamma B^2 = 0.5007 $$

Looking at the synchrotron energy $E = h\nu_{sync} = 4keV$, the frequency $\nu_{sync} = \frac{E}{h} = 9.69\times10^{17}\ Hz$. Equating this frequency to the synchrotron frequency gives

$$ \nu_{sync} = \frac{3q}{4\pi mc} \gamma^2 B = \frac{3e}{4\pi m_e c} \gamma^2 B $$

Substituting the appropriate values, the equation can be solved for $\gamma^2 B$:

$$ 9.69\times10^{17} = \frac{3(4.8\times10^{-10})}{4\pi (9.1\times10^{-28})(3\times10^{10})} \gamma^2 B $$

$$ \gamma^2 B = 2.310\times10^{11} $$

These two equations for $\gamma B^2$ and $\gamma^2 B$ can be combined to solve for $\gamma$ and $B$ individually.

$$ \frac{\gamma^2 B}{(\gamma B^2)^2} = \frac{1}{B^3} = 9.215 $$

$$ B = 1.03\times10^{-4}G = 103\ \mu G $$

$$ \frac{(\gamma^2 B)^2}{\gamma B^2} = \gamma^3 = 1.065\times10^{23} $$

$$ \gamma = 4.74\times10^7 $$

This gives an electron energy (in eV) of

$$ E_e = \gamma m_e c^2 = (4.74\times10^7)(0.511\ MeV) = 2.37\times10^{13}\ eV $$

***

b. For synchrotron radiation at $h \nu_{sync} = 1\ keV$, the value of $\gamma$ is different for the gyrating electrons. This leads to a differend rim width.

Under the same magnetic field as part a,

$$ \nu_{sync} = \frac{E}{h} = \frac{3q}{4\pi mc} \gamma^2 B $$

$$ \frac{1000(1.6\times10^{-12})}{6.6\times10^{-27}} = \frac{3(4.8\times10^{-10})}{4\pi (9.1\times10^{-28})(3\times10^{10})} \left(1.03\times10^{-4}\right) \gamma^2 $$

$$ \gamma = 2.368\times10^7 $$

This new value of $\gamma$ can be used to find $d$ by using the electron velocity $v_e$ and cooling time $t_c$

$$ d = v_e \frac{m_e^3 c^5}{e^4} \frac{1}{\gamma B^2} = 4500\times10^5 \frac{(9.1\times10^{-28})^3 (3\times10^{10})^5}{(4.8\times10^{-10})^4} \frac{1}{(2.368\times10^7) (1.03\times10^{-4})^2} $$

$$ d = 0.2\ pc $$

(This could be seen from the original equation, as $\gamma$ decreased by roughly a factor of 2.)

If the width of the shock rim does not change with photon energy, then the rim is independent of synchrotron frequency, and therefore electron energy.  A possible alternative explanation is that the shock rim is a region where photon emission is stimulated in a way independent of initial energy; for example, the rim may be created by bremsstrahlung radiation from charged particles at the shock front interacting with the ISM.
