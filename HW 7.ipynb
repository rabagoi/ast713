a. Impact parameter $b$, relation to inclination angle $i$ and and transit suration $t_{dur}$

For a planetary system at an angle $i$ to the plane of the sky, a right triangle can be formed using the star, the orbit of the planet, and the sky plane. This triangle is easiest seen if the viewed along the line created by the intersection of the planet's orbital plane and the sky plane.

The right triangle has a hypotenuse $a$ and acute angle $i$ at the vertex corresponding to the star's location. The horizontal distance is the impact parameter $b$, which is also the projection of the planet's semi-major azis onto the sky plane. Measured in units of length,

$a_x = br_*$

Thus, the inclination angle and impact parameter are related by

$$ \cos i = \frac{a_x}{a} = \frac{br_*}{a}$$

The impact parameter also determines the duration of the transit, $t_{dur}$. A transit causes the planet to traverse a chord of the star's face, the length of which is

$$ L_{transit} = 2\sqrt{1-b^2} r_*$$

(where the equation is multiplied by $r_*$ to keep the result in units of length).  Assuming a circular orbit for the planet, the transit duration can be expressed as

$$ t_{dur} = \frac{d_{trans}}{v_p} = \frac{d_{trans}}{\frac{2 \pi a}{\tau}} $$

$$ t_{dur} = \frac{2 \sqrt{1-b^2}r_*}{\frac{2 \pi a}{\tau}} $$

$$ t_{dur} = \frac{r_* \tau}{\pi a} \sqrt{1-b^2}$$

where $a$ is the planet's semi-major axis and $\tau$ is the planet's orbital period.

b. Probability of transit

A transit can occur along angles slightly misaligned from the orbital plane, up to a grazing transit, where the observer's line of sight is an inner common tangent to the star and the planet. This line creates a right triangle with the planet's semi-major axis as the horizontal, with right triangle legs $a$ and $R_* + R_p$.

Using the small angle approximation, the largest angle $\theta_{max}$ at which a grazing transit occurs is

$$ \theta_{max} \approx \tan \theta_{max} = \frac{R_* + R_p}{a} $$

The transits can be viewed on either side of the orbital plane, in any of the radial directions pointing towards the star, so the total angular area that can view a planetary transit is

$$ A_{transit} = 2 \pi\ \cdot 2\ \cdot \theta_{max} = 4\pi \frac{R_* + R_p}{a}$$

The probability of a transit can be found by dividing this value by the full $4\pi$ steradians of the sky.

$$ P_{transit} = \frac{A_{transit}}{4\pi} =  \frac{R_* + R_p}{a}$$

c. Kepler-93 system: $M_* = 0.91 M_\odot ,\ r_* = 0.92 r_\odot ,\ T_E = 5700\ K ,\ A_p = 0.3$

Radial velocity measurements indicate variations with an amplitude of $v_{obs} = 1.6\ m\ s^{-1}$.

Using the data in Figure 6.2, the transit period, duration, and depth are roughly

$ \tau \approx 4.6\ days $

$ t_{trans} \approx 0.03 \tau = 0.138\ days$

$ \frac{\Delta f}{f} \approx 0.00025 $

The planet's semi-major axis can be found using Kepler's Third Law

$$ \frac{a^3}{\tau^2} = \frac{GM}{4 \pi^2} $$

$$ a = \left( \frac{(6.7\times10^{-8}) \cdot {0.91(2\times10^{33})}}{4\pi^2} \cdot 4.6(86400\ s) \right)^{\frac{1}{3}} $$

$$ a = 7.8724\times10^{11} cm = 0.05248\ AU $$

The orbital inclination can be found by combining the equations derived in part a).

$$ t_{dur} = \frac{r_* \tau}{\pi a} \sqrt{1-b^2} = \frac{r_* \tau}{\pi a} \sqrt{1-\left( \frac{a}{r_*} \cos i \right)^2}$$

$$ \cos i = \frac{r_*}{a} \sqrt{1 - \left( \pi \frac{a\ t_{dur}}{r_* \tau} \right)^2} $$

$$ \cos i = 0.0468 $$

$$ i = 87.3173^\circ $$

The change in flux of the starlight can be used to find the planet's radius.

$$ \frac{\Delta f}{f} = \left( \frac{r_p}{r_*} \right)^2 $$

$$ r_p = r_* \sqrt{\frac{\Delta f}{f}} = 0.92R_\odot \sqrt{0.00025} $$

$$ r_p = 1.018\times10^9\ cm = 1.591 r_\oplus $$

The radial velocity measurements can be used to find the mass of the planet using the relation between mass and velocity for binaries. For a planet-star system, the star is much more massive than the planet, so $M_* \gg M_p$, so the equation simplifies to

$$ \frac{M_p^3}{(M_p + M_*)^2} \sin^3 i = \frac{\tau |v_{1, obs}|^3}{2 \pi G} \longrightarrow \frac{M_p^3}{M_*^2} \sin^3 i = \frac{\tau |v_{1, obs}|^3}{2 \pi G}$$

$$ M_p \sin i = \left( \frac{\tau}{2 \pi G} \right)^{\frac{1}{3}} |v_{1, obs}|\ M_*^\frac{2}{3} $$

$$ M_p = 2.3458\times10^{28}\ g = 3.9097 M_\oplus $$

This puts the density of the planet at

$$ \rho_p = \frac{M_p}{\frac{4}{3} \pi r_p^3} $$

$$ \rho_p = 5.304 \frac{g}{cm^3} $$

Finally, the temperature of the planet can be found using the star's temperature, the planet's albedo, and the equation for thermal equilibrium.

$$ T_p = 2^{-\frac{1}{2}} \left( \frac{r_*}{a} \right)^{\frac{1}{2}} (1-A)^{\frac{1}{4}} T_* $$

$$ T_p = 2^{-\frac{1}{2}} \left( \frac{0.92(7.0\times10^{10})}{7.8724\times10^{11}} \right)^{\frac{1}{2}} (1-0.3)^{\frac{1}{4}} (5700\ K) $$

$$ T_p = 1054\ K $$

In summary, this planet orbits very close to its star, and is a super-earth of roughly the same density as the Earth.
